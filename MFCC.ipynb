# Preprocessing Audio data for DeepLearning

```python
import librosa,librosa.display
```

```python
file="blues.00000.wav"
```

```python
#wave form
signal,sr=librosa.load(file,sr=22050) # sr*T ->22050 *30
```

```python
# fft ->spectram
# stft ->specrogram
# MFCCs
```

```python
#librosa.display.waveplot(signal,sr=sr)
import librosa



# Plot the waveform using waveshow()
librosa.display.waveshow(signal, sr=sr)
import matplotlib.pyplot as plt
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
import numpy as np
fft=np.fft.fft(signal)
magnitude=np.abs(fft)
frequency=np.linspace(0,sr,len(magnitude))
left_frequency=frequency[:int(len(frequency)/2)]
left_magnitude=magnitude[:int(len(frequency)/2)]
plt.plot(left_frequency,left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.show()
```

```python
# stft Spectrogram
n_fft=2048
hop_length=512

stft=librosa.core.stft(signal,hop_length=hop_length,n_fft=n_fft)
spectrogram=np.abs(stft)

log_spectrogram=librosa.amplitude_to_db(spectrogram)
librosa.display.specshow(log_spectrogram, sr=sr ,hop_length=hop_length)
import matplotlib.pyplot as plt



plt.xlabel("Time")
plt.ylabel("Frequency")
plt.show()
#plt.colorbar()
plt.show()

```

```python
#MFFCs
#MFFCs=librosa.feature.mfcc(signal,n_fft=n_fft , hop_length=hop_length,n_mfcc=13)
import librosa

# Load the audio signal
signal, sample_rate = librosa.load("blues.00000.wav")

# Calculate the MFCCs
MFCC = librosa.feature.mfcc(y=signal, n_fft=2048, hop_length=512, n_mfcc=13)
librosa.display.specshow(MFCC, sr=sr ,hop_length=hop_length)

# Print the MFCCs
print(MFCC)
plt.xlabel("Time")
plt.ylabel("MFCC")
plt.show()
#plt.colorbar()
plt.show()
```
